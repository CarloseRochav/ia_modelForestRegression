# I. Analisis Exploratorio Visual

### 1. Importamos las librerías necesarias:
- Pandas: para cargar y manipular los datos
- Matplotlib: para graficar
- Seaborn: para visualizaciones avanzadas

```python
# Importar librerías
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### 2. Cargamos el conjunto de datos CSV en un DataFrame de Pandas con pd.read_csv(). 
### Esto nos da una representación de la data en forma de tabla en Python.

```python
# Cargar datos
df = pd.read_csv("../tableClaude.csv")
```

### 3. Graficamos líneas para visualizar la tendencia de cada variable a través del tiempo:
- Usamos df.plot() y pasamos los nombres de columnas para generar un gráfico de líneas.
- Como cada fila es un periodo, vemos la evolución de esa variable en el tiempo.

```python
# Gráfico de líneas  
#df.plot(y=['sem1','sem2','sem3',...], kind='line') 
df.plot(y=['sem1','sem2','sem3','sem4','sem5','sem6','sem7','sem8','sem9'], kind='line') 
plt.show()
```

### 4. Calculamos la matriz de correlación entre todas las variables con df.corr(). Esto muestra las dependencias lineales entre ellas.

### 5. Usamos Seaborn para visualizar la matriz de correlación como un heatmap con sns.heatmap().
- Los valores cercanos a 1 o -1 indican alta correlación.

```python
# Matriz de correlación
corr_matrix = df.corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True)
plt.show()

# Otros gráficos exploratorios
# df.hist()
# df.boxplot()
# etc
```
